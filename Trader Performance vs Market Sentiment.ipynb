**Part B – Analysis**

**Creating Features for Analysis**

```python
merged_df["Winning Trade"] = merged_df["Closed PnL"] > 0
```

```python
def position_type(direction):
    direction = str(direction).lower()

    if "long" in direction or direction == "buy":
        return "Long"
    elif "short" in direction or direction == "sell":
        return "Short"
    else:
        return "Other"

merged_df["Position Type"] = merged_df["Direction"].apply(position_type)
```

**Question 1: Does trader performance differ between Fear and Greed market conditions?**

```python
pnl_by_sentiment = (
    merged_df.groupby("classification")["Closed PnL"]
    .mean()
    .reset_index()
    .round(2)
)

pnl_by_sentiment
```

```python
winrate_by_sentiment = (
    merged_df.groupby("classification")["Winning Trade"]
    .mean()
    .reset_index()
    .round(2)
)

winrate_by_sentiment["Winning Trade"] = winrate_by_sentiment["Winning Trade"] * 100

winrate_by_sentiment.rename(
    columns={"Winning Trade": "Win Rate (%)"},
    inplace=True
)

winrate_by_sentiment
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(
    pnl_by_sentiment["classification"],
    pnl_by_sentiment["Closed PnL"]
)

plt.title("Average Closed PnL by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average Closed PnL")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    winrate_by_sentiment["classification"],
    winrate_by_sentiment["Win Rate (%)"]
)

plt.title("Win Rate by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Win Rate (%)")

plt.show()
```

```python
drawdown_proxy = (
    merged_df[merged_df["Closed PnL"] < 0]
    .groupby("classification")["Closed PnL"]
    .mean()
    .reset_index()
)

drawdown_proxy.rename(
    columns={"Closed PnL": "Average Loss (Drawdown Proxy)"},
    inplace=True
)

drawdown_proxy
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(
    drawdown_proxy["classification"],
    drawdown_proxy["Average Loss (Drawdown Proxy)"]
)

plt.title("Average Loss (Drawdown Proxy) by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average Loss")

plt.show()
```

**Insight 1**

**Traders achieved the highest average Closed PnL during Extreme Greed market conditions (67.89), indicating that trading performance was strongest during periods of high market optimism.**

**Insight 2**

**The highest win rate (46.49%) was also observed during Extreme Greed, suggesting traders executed more successful trades when market sentiment was strongly positive.**

**Insight 3**

**Extreme Fear showed the lowest win rate (37.06%), indicating that traders generally found it more challenging to generate profitable trades during highly fearful market conditions.**

**Insight 4**

**The largest average loss (drawdown proxy) occurred during Extreme Fear market conditions (−257.10), while Extreme Greed had the smallest average loss (−119.92). This suggests that traders experienced higher downside risk during fearful markets and comparatively lower downside risk during strongly bullish market conditions.**

**Question 2: Do traders change behaviour based on market sentiment?**

```python
trade_size_by_sentiment = (
    merged_df.groupby("classification")["Size USD"]
    .mean()
    .reset_index()
    .round(2)
)

trade_size_by_sentiment
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    trade_size_by_sentiment["classification"],
    trade_size_by_sentiment["Size USD"]
)

plt.title("Average Trade Size by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Average Trade Size (USD)")

plt.show()
```

```python
trades_by_sentiment = (
    merged_df.groupby("classification")
    .size()
    .reset_index(name="Number of Trades")
)

trades_by_sentiment
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    trades_by_sentiment["classification"],
    trades_by_sentiment["Number of Trades"]
)

plt.title("Number of Trades by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Number of Trades")

plt.show()
```

```python
position_by_sentiment = (
    merged_df.groupby(["classification","Position Type"])
    .size()
    .unstack(fill_value=0)
)

position_by_sentiment
```

```python
position_by_sentiment.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Long vs Short Positions by Market Sentiment")
plt.xlabel("Market Sentiment")
plt.ylabel("Number of Positions")
plt.xticks(rotation=45)

plt.show()
```

**Insight 1 – Average Trade Size**

**Traders placed the largest average trade size during Fear market conditions (approximately USD 7,816). This suggests that traders were willing to take larger positions during fearful markets, possibly expecting price reversals.**

**Insight 2 – Trading Activity**

**The highest number of trades occurred during Fear (61,837 trades), followed by Greed (50,303 trades). This indicates that trading activity increased during volatile market conditions.**

**Insight 3 – Long vs Short Behaviour**

**Traders showed a long position bias during Fear and Neutral market conditions, while short positions exceeded long positions during Greed and Extreme Greed, indicating that trader positioning changes with market sentiment.**

**Question 3: Trader Segmentation**

*Segment 1: Profitable vs Loss-Making Traders*

```python
trader_segments = merged_df.groupby("Account").agg(
    Total_PnL=("Closed PnL", "sum"),
    Total_Trades=("Account", "count"),
    Average_Trade_Size=("Size USD", "mean")
).reset_index()

trader_segments.head()
```

```python
trader_segments["Trader Segment"] = trader_segments["Total_PnL"].apply(
    lambda x: "Profitable" if x > 0 else "Loss Making"
)

trader_segments["Trader Segment"].value_counts()
```

*Segment 2: Frequent vs Infrequent Traders*

```python
median_trades = trader_segments["Total_Trades"].median()

trader_segments["Trading Frequency"] = trader_segments["Total_Trades"].apply(
    lambda x: "Frequent Trader" if x >= median_trades else "Infrequent Trader"
)

trader_segments["Trading Frequency"].value_counts()
```

*Segment 3: Consistent vs Inconsistent Winners*

```python
consistency = (
    merged_df.groupby("Account")["Winning Trade"]
    .mean()
    .reset_index()
)

consistency["Win Rate (%)"] = consistency["Winning Trade"] * 100

consistency["Consistency"] = consistency["Win Rate (%)"].apply(
    lambda x: "Consistent Winner" if x >= 50 else "Inconsistent Winner"
)

consistency["Consistency"].value_counts()
```

**Trader Segmentation Insights**

**Insight 1**

**Traders were segmented based on overall profitability. Out of 32 traders, 29 were profitable while 3 were loss-making.**

**Insight 2**

**Traders were also segmented based on trading frequency using the median number of trades. This resulted in 16 Frequent Traders and 16 Infrequent Traders, providing a balanced comparison.**

**Insight 3**

**Traders were further segmented based on win rate. Only 3 traders achieved a win rate of 50% or more and were classified as Consistent Winners, while 29 traders were classified as Inconsistent Winners.**

**Strategy Recommendations**

**Strategy Recommendation 1**

**Reduce position size during Extreme Fear market conditions, as traders achieved the lowest win rate (37.06%) during these periods. A more cautious approach may help reduce losses.**

**Strategy Recommendation 2**

**Increase trading activity during Extreme Greed only with proper risk management, since this sentiment showed both the highest average Closed PnL and the highest win rate (46.49%).**

**Conclusion**

**1. Extreme Greed market conditions showed the highest average Closed PnL and win rate.**

**2. Trader behaviour varied across different market sentiments, especially in trade size, trading activity, and position type.**

**3. Trader segmentation identified profitable, frequent, and consistent traders, helping derive practical trading recommendations.**
